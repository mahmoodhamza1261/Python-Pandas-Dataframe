 Handle Missing Values: Create a DataFrame with some missing values:
Name | Age | Salary
---------------------
Alice | 30 | 50000
Bob | NaN | 60000
Carol | 29 | NaN
Dave | 35 | 70000
Fill the missing values in the Age column with the mean age and in the Salary column with the
median salary.

```python
df3=pd.DataFrame({'Name':['Alice','Bob','Carol','Dave'],'Age':[30,np.nan,29,35],'Salary':[50000,60000,np.nan,70000]})
df3
```

```python
df3['Age'].fillna(df3['Age'].mean(),inplace=True)
df3['Salary'].fillna(df3['Salary'].median(),inplace=True)
df3
```

Combine DataFrames: Create two DataFrames:
df1:
Name | Department
---------------------
Alice | HR
Bob | Engineering
df2:
Name | Salary
------------------
Carol | 55000
Dave | 70000
Combine df1 and df2 based on the Name column, handling the missing values as needed.

```python
df1=pd.DataFrame({'Name':['Alice','Bob'],'Department':['HR','Engineering']})
df2=pd.DataFrame({'Name':['Carol','Dave'],'Salary':[55000,70000]})
merged=df1.merge(df2,on='Name',how='outer')
merged['Salary'].mean()
```

```python
merged['Department'].fillna('Unknown',inplace=True)
merged['Salary'].fillna(merged['Salary'].mean(),inplace=True)
merged
```
